```python
import zipfile
import pandas as pd
import glob
from google.colab import drive


drive.mount('/content/drive')


zip_path = '/content/drive/MyDrive/archive (2).zip'  # Path to your ZIP in Drive
extract_path = '/content/drive/MyDrive/stock_data'   # Folder where CSVs will be extracted


with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Extraction complete!")


all_files = glob.glob(extract_path + "/*.csv")
li = []

for filename in all_files:
    df_temp = pd.read_csv(filename)
    df_temp['Name'] = filename.split('/')[-1].replace('.csv','')  # stock ticker
    li.append(df_temp)

df = pd.concat(li, axis=0, ignore_index=True)

print("Combined dataset shape:", df.shape)
df.head()
```

```python

df['Date'] = pd.to_datetime(df['Date'])


df_pivot = df.pivot_table(index='Date', columns='Name', values='Close', aggfunc='mean')


df_pivot.fillna(method='ffill', inplace=True)


portfolio = df_pivot.sample(n=10, axis=1, random_state=42)

print("Sampled portfolio shape:", portfolio.shape)
portfolio.head()
```

```python

returns = portfolio.pct_change().dropna()
returns.head()
```

```python

weights = np.array([1/portfolio.shape[1]]*portfolio.shape[1])

portfolio_returns = returns.dot(weights)
portfolio_returns.head()
```

```python

VaR_95 = np.percentile(portfolio_returns, 5)
print(f"Historical 95% VaR: {VaR_95:.4f}")
```

```python
num_simulations = 100000
mean_ret = portfolio_returns.mean()
std_ret = portfolio_returns.std()

simulated_returns = np.random.normal(mean_ret, std_ret, num_simulations)


MC_VaR_95 = np.percentile(simulated_returns, 5)
print(f"Monte Carlo 95% VaR: {MC_VaR_95:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(portfolio_returns, bins=50, kde=True, color='blue', label='Observed Returns')
sns.histplot(simulated_returns, bins=50, kde=True, color='red', label='Simulated Returns', alpha=0.5)
plt.axvline(VaR_95, color='blue', linestyle='--', label='Historical VaR 95%')
plt.axvline(MC_VaR_95, color='red', linestyle='--', label='Monte Carlo VaR 95%')
plt.title("Portfolio Returns Distribution & VaR")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
